```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
data = pd.read_csv("/content/drive/MyDrive/Sample99.csv") 
data
```

```python
x=np.array(data.Param1).reshape(-1, 1)
y=np.array(data.Param2).reshape(-1, 1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, random_state = 12, test_size = 30)
print(y.shape, y_train.shape, y_test.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(x_train, y_train)
```

```python
print(model.coef_, model.intercept_)
```

```python
y_predict= model.predict(x_test)
y_predict
```

```python
metrics.mean_squared_error(y_test, y_predict, squared=False)
```

```python
y.mean()
```

```python
metrics.r2_score(y_test, y_predict)
```
