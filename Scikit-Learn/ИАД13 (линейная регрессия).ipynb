```python
from google.colab import drive
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import scipy.stats as st
from sklearn.linear_model import LinearRegression
from sklearn import metrics
```

```python
drive.mount("/content/drive")
```

```python
data = pd.read_csv("/content/drive/MyDrive/Sample99.csv")
data
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(data.Param1, data.Param2)
```

```python
st.pearsonr(data.Param1, data.Param2)[0]
```

```python
model = LinearRegression()
```

```python
model.fit(np.array(data.Param1).reshape(-1, 1), np.array(data.Param2).reshape(-1, 1))
```

```python
model.coef_
```

```python
model.intercept_
```

```python
Param2_predict = model.predict(np.array(data.Param1).reshape(-1, 1))
Param2_predict
```

```python
plt.figure(figsize=(14, 8))
plt.scatter(data.Param1, data.Param2)
plt.plot(data.Param1, Param2_predict, "black")
```

```python
metrics.mean_squared_error(data.Param2, Param2_predict, squared=False)
```
