```python
import pandas as pd
```

```python
arr = [['No','No','Yes','Yes'],['Lowest','Highest','Lowest','Highest']]
index1 = pd.MultiIndex.from_arrays(arr,names=['Picture','Rank'])
index1
```

```python
ser = pd.Series(['Two','Ace','Jack','King'], index = index1)
ser
```

```python
data = pd.DataFrame({'Spades': ['2♠', 'A♠', 'J♠', 'K♠'], 
                     'Diamonds': ['2♦', 'A♦', 'J♦', 'K♦'], 
                     'Hearts': ['2♥', 'A♥', 'J♥', 'K♥'], 
                     'Clubs': ['2♣', 'A♣', 'J♣', 'K♣']}, index=index1)
data
```

```python
tup = [('Real','Positive'),('Real','Negative'),('Imaginary','Positive'),('Imaginary','Negative')]
index2 = pd.MultiIndex.from_tuples(tup, names = ('Axis','Direction'))
index2
```

```python
data2 = pd.DataFrame({'One':["1","-1","i","-i"]},index = index2)
data2
```

```python
note = ["C", "D", "E", "F", "G", "A", "B"]
pitch = ['Natural', 'Sharp',]
index3 = pd.MultiIndex.from_product([note, pitch], names=['Note','Pitch'])
index3
```

```python
data3 = pd.DataFrame({'Symbol': ["C/B#", "C#/Db", "D", "D#/Eb", "E/Fb", "F/E#", "F/E#", "F#/Gb", "G", "G#/Ab", "A", "A#/Bb", "B/Cb", "C/B#"]},index = index3)
data3
```
