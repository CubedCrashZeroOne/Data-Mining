```python
arr = [["two", "two", "three", "three", "four", "four"], ["squared", "cubed", "squared", "cubed", "squared", "cubed"]]
index = pd.MultiIndex.from_arrays(arr, names = ("number", "power"))
math = pd.DataFrame({"result": [4, 8, 9, 27, 16, 64]}, index=index)
math
```
