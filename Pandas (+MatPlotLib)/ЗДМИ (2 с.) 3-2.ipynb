Загрузить данные из csv файла (asia_2000_2015.csv) в датафрейм df_as

```python
df_as = pd.read_csv("/content/drive/MyDrive/concat2/AS00-15.csv")
df_as
```

Создайте датафрейм (на основе словаря) df_areas для хранения
информации о площадях разных стран

image.png

```python
df_areas = pd.DataFrame({"Area": [17100000, 100210, 20770, 3287000], "Country": ["Russia", "Korea", "Israel", "India"]})
df_areas
```

Используя pd.merge() с ‘on’ кл.словом выполнить объединение
датафреймов для получения следующего результата

image.png

```python
pd.merge(df_areas, df_as, on="Country")
```

Создайте датафрейм (на основе словаря) df_population для хранения
информации о численности населения разных стран

image.png

```python
df_pop = pd.DataFrame({"Country": ["Russia", "Korea", "Israel", "India"], "Population": [144000000, 76500000, 8547000, 1318700000]})
df_pop
```

```python
pd.merge(df_areas, df_pop, on="Country")
```

Используя pd.merge() с ‘on’ кл.словом выполнить объединение
датафреймов для получения следующего результата

image.png

```python
pd.merge(pd.merge(df_areas,  df_as, on="Country"), df_pop, on="Country")
```
