```python
data = pd.read_csv("/content/drive/MyDrive/concat3/final.csv", index_col=0)
data
```

1. Select subsets of data (final dataset (US States Data) from prev. lab)
based on ‘state\region’ column. Output the number of observation in
each group.

```python
data.groupby("state/region")["state/region"].count()
```

2. Calculate mean value of population for each state.

```python
data.groupby("state/region")["population"].mean()
```

3. Output the min value of population for each state and year.

```python
data.groupby(["state/region", "year"])["population"].min()
```

4. Select subsets of data based on ‘year’ column. Calculate sum of
population for each year.

```python
data.groupby("year")["population"].sum()
```

5. By using aggregate() method output median, std and max value for
each state

```python
gb_pop = data.groupby("state/region").population
gb_pop
```

```python
gb_agg = gb_pop.agg(["median", "std", "max"])
gb_agg
```

6. By using filter() method output 10 1st groups in which the standard deviation of population is larger than X=3.237715e+06.

```python
gb_std = gb_agg.filter(items=["std"])
gb_std[gb_std["std"] > 3.237715e+06].head(10)
```
