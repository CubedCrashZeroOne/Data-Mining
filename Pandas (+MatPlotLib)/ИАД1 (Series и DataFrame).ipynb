```python
import pandas as pd
```

1. Создать Series

```python
num_list = [1, 2, 3, 4, 5]
num_ser = pd.Series(num_list)
num_ser
```

2. Переопределить индексы

```python
num_ser.index = ["zero", "one", "two", "three", "four"]
num_ser
```

3. Выбрать конкретные элементы

```python
print(num_ser["zero"])
print(num_ser["three"])
```

4. Создать DataFrame

```python
letters = {"a": ["a", "aa", "aaa"], "b": ["b", "bb", "bbb"], "x":["x", "xx", "xxx"], "y": ["y", "yy", "yyy"]}
let_df = pd.DataFrame(letters)
let_df
```

5. Переопределить индексы

```python
let_df.index = ["single", "double", "triple"]
let_df
```

6. Выбрать конкретные строки

```python
let_df.loc(0)["single"]
```

```python
let_df.loc(0)["double"]
```

```python
let_df.loc(0)["triple"]
```
