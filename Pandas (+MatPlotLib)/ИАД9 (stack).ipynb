```python
import pandas as pd
```

```python
data = pd.DataFrame([[0, 32], [36, 92], [100, 212]], index=["Freezing water", "Human body", "Boiling water"], columns=["°C", "°F"])
data
```

```python
sdata = data.stack()
sdata
```

```python
sdata.unstack()
```

```python
cols = pd.MultiIndex.from_tuples((("Temperature", "°C"), ("Temperature", "°F")))
data = pd.DataFrame([[0, 32], [36, 92], [100, 212]], index=["Freezing water", "Human body", "Boiling water"], columns=cols)
data
```

```python
sdata = data.stack()
sdata
```

```python
sdata.unstack()
```

```python
sdata = data.stack(level=0)
sdata
```

```python
sdata.unstack()
```

```python
sdata.unstack(level=0)
```

```python
sdata.unstack(level=1)
```
