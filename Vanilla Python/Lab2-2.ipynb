Создайте словарь студентов, которые участвуют в опросе по поводу
любимого языка программирования (не менее 10 элементов): фамилия –
язык программирования. Создайте отдельный список студентов (только
фамилии, не менее 15), которые желают участвовать в опросе. Пусть в этом
списке присутствуют несколько фамилия из первого словаря. Напишите
функцию, которая перебирает фамилии из списка следующим образом:
если студент проходил опрос, то его благодарят за участие, и переходят к
следующему, если не проходил, то спрашивают о любимом языке
программирования и заносят соответствующую информацию в словарь.
После проверки всего списка вывести содержимое словаря в
отсортированном по алфавиту порядке (по фамилиям студента).

```python
dic = dict(Гусин='C#', Шабай='Python', Макогон='Javascript', Андреев='C#', Лашевич='Java', Попов='Javascript', Дьяченко='C++', Дрига='Javascript', Пшеник='C++', Сас='Javascript')

list = ["Гусин", "Гордиенко", "Ласковец", "Андриевский", "Лашевич", "Сас", "Попов", "Леник", "Пакунов", "Макогон", "Кадин", "Крошин", "Козырев", "Шабай", "Косько"]

for a in list:
  print(a)
  if a in dic:
    print("Благодарим за участие!")
  else:
    print("Выберите язык:")
    dic[a] = input()

print('\nСписок:\n')

for b in sorted(dic):
  print(b + ": " + dic[b])
```
