4. Написать функцию is_year_leap, принимающую 1 аргумент — год, и возвращающую True, если год високосный, и False иначе.


```python
def is_leap(year):
  if (year % 100 == 0 and year % 400 != 0) or year % 4 != 0:
    return False
  else:
    return True
```
