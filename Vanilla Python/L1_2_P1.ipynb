6. Напишите функцию Python для подсчета количества комнат, площадь
которых больше 10 квадратных метров, и распечатайте их названия.

```python
#task 6
def count_more_than_ten(elemlist):
  count = 0
  for a in elemlist:
    if a[1] > 10:
      count += 1
      print(a[0])
  return count

print('count: ' + str(count_more_than_ten(list)))
```

7. Напишите функцию Python для вывода имен комнаты, где длина имени (в
виде строки) равна 8 или больше и заканчивается «комнатой»
(используйте срезы списков с конца)

```python
# task 7

def print_room_names(elemlist):
  for a in elemlist:
    if len(a[0]) >= 8 and a[0].endswith('room'):
      print(a[0])
  
print_room_names(list)
```

8. Посчитайте номер комнаты, где площадь комнаты больше 10 метров и
выведите их названия.

```python
# task 8

def room_index(elemlist):
  for a in elemlist:
    if a[1] >= 10:
      print(str(elemlist.index(a)) + ': ' + a[0])


room_index(list)
```

9. Вы ошибочно записали повторяющуюся информацию в списке (о коридоре
и его площади). Но что, если это не единственный случай! Напишите
функцию Python для удаления дубликатов (повторяющееся название
комнаты и ее площадь) из списка.

```python
# task 9

list = [['hallway', hall], ['kitchen', kit], ['bedroom', bed], ['hallway', hall], ['living room', liv], ['bedroom', bed], ['bathroom', bath], ['hallway', hall]]

for a in list:
  count = list.count(a)
  if count > 1:
    for b in range(1, count):
      list.remove(a) 

print(list)
```
