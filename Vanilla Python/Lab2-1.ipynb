1. Создайте словарь для хранения информации о товаре: название, категория,
краткое описание, производитель, цена. Выведите каждый фрагмент
информации, хранящейся в словаре.

```python
dic = dict(name = "Sprite", category = "Drink", description = "A 2-litre bottle of a delicious lemon-flavoured soda", manufacturer = "Coca Cola", price = 38.50)


for b in dic:
    print(f"{b}: {dic[b]}")
```
