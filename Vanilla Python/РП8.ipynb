4. Создайте класс StringVar для работы со строковым типом данных, содержащий методы set и get. Метод set служит для изменения содержимого строки, get – для получения содержимого строки. Создайте объект типа StringVar и протестируйте его методы.


```python
class StringVar:
  def __init__(self, stringinput):
    self.string = stringinput
  def set_str(self, val):
    self.string = val
  def get_str(self):
    return self.string
```

```python
obj = StringVar("aoaoao")
obj.get_str()
```

```python
obj.set_str("oaoaoa")
obj.get_str()
```
