7. Создать матрицу a x b с 5 внутри, и 1 на границах

```python
matrix_ab = np.zeros((a, b), dtype = 'int32')

matrix_ab[::a-1] = 1
matrix_ab[::,::b-1] = 1
matrix_ab[1:a-1, 1:b-1] = 5


matrix_ab
```
