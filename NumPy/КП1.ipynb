10. Прочитати з файлу частину стовбця book_id відповідно до частини
user_id, сформувати NumPy масив В_id. Для кожної книги з В_id на основі
даних з CRT розрахувати загальну кількість хвилин, яку на неї витратили
клієнти у першому місяці.

```python
b_id = file_array[a_from:a_to, 0]
b_id
```

```python
uniq_b_id = np.unique(b_id) 
uniq_b_id
```

```python
length = len(uniq_b_id)
minutes = np.zeros(length, dtype="int32")
for i in range(length):
  minutes[i] = np.sum(crt[b_id == uniq_b_id[i], 0])

minutes
```
